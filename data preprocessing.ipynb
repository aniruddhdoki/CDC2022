```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
data_url = "https://www.ncei.noaa.gov/pub/data/swdi/stormevents/csvfiles/"
```

```python
r = requests.get(data_url)
html = BeautifulSoup(r.content, 'html5lib')
links = html.find_all('a')
links = [data_url + link['href'] for link in links if link['href'].endswith('.csv.gz')]
links
```

```python
for link in links:
  file_name = '/content/drive/MyDrive/cdc2022/data/' + link.split('/')[-1]
  print(f'downloading {file_name}')
  request = requests.get(link) 
  with open(file_name, 'wb') as file: 
    file.write(request.content)
```

```python
root_dir = '/content/drive/MyDrive/cdc2022/data'
details_df = []
fatalities_df = []
locations_df = []
files = os.scandir(root_dir)

def read_csv(root_dir, file_name):
  return pd.read_csv(root_dir + '/' + file_name, compression="gzip")

for csv in files:
  csv = csv.name
  if "details" in csv:
    details_df.append(read_csv(root_dir, csv))
  elif "fatalities" in csv:
    fatalities_df.append(read_csv(root_dir, csv))
  elif "locations" in csv:
    locations_df.append(read_csv(root_dir, csv))
  else:
    print(f'error on {csv}')
details_df = pd.concat(details_df)
details_df.to_csv('/content/drive/MyDrive/cdc2022/details.csv')
fatalities_df = pd.concat(fatalities_df).to_csv('/content/drive/MyDrive/cdc2022/fatalities.csv')
locations_df = pd.concat(locations_df).to_csv('/content/drive/MyDrive/cdc2022/locations.csv')
```
